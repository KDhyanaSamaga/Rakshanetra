```python
# scan_labels.py
import glob, os

labels_dir = '/content/Rakshanetra/yolov5/labels'  # set to your labels dir
num_classes = 3  # expected classes: 0..2

max_idx = -1
bad_files = []  # (file, class_index, line)
for path in glob.glob(os.path.join(labels_dir, '*.txt')):
    with open(path, 'r') as f:
        for ln, line in enumerate(f, start=1):
            s = line.strip()
            if not s:
                continue
            parts = s.split()
            try:
                cls = int(float(parts[0]))
            except:
                print(f"Could not parse class in {path}:{ln} -> {s}")
                continue
            if cls > max_idx:
                max_idx = cls
            if cls >= num_classes:
                bad_files.append((path, cls, ln, s))

print(f"Max class index found across labels: {max_idx}")
print(f"Total files/lines with index >= {num_classes}: {len(bad_files)}")
if bad_files:
    print("Examples (up to 20):")
    for i, (p, cls, ln, s) in enumerate(bad_files[:20]):
        print(f"{i+1}. {p} (line {ln}) -> class {cls} | '{s}'")
else:
    print("No invalid indices found. Labels look consistent.")
```

```python
# ONE-CELL DIAGNOSTIC: search labels, show data.yaml, inspect checkpoint
import os, glob, yaml, torch
from pathlib import Path
pd = Path

print("==== 1) Searching for label .txt files (recursive) ====")
root = '/content'   # adjust if your project root is elsewhere
label_files = []
for p in pd(root).rglob('*.txt'):
    # Heuristic: YOLO label files usually live under a 'labels' folder or have numeric contents per line
    if 'label' in str(p).lower() or 'labels' in str(p).lower() or True:
        label_files.append(str(p))
label_files = sorted(label_files)
print(f"Found {len(label_files)} .txt files under {root} (showing up to 50 matches):")
for f in label_files[:50]:
    print(" -", f)

# If many .txt files, show a quick peek of those that look like YOLO labels (numeric first token)
print("\nPeeking into likely label files (first 50):")
count = 0
for f in label_files[:200]:
    try:
        with open(f, 'r') as fh:
            lines = [ln.strip() for ln in fh.readlines() if ln.strip()]
        # consider it probably a YOLO label if lines exist and first token is numeric
        if lines:
            first = lines[0].split()[0]
            is_num = False
            try:
                float(first); is_num = True
            except:
                is_num = False
            if is_num:
                print(f"\n-- {f} (showing up to 10 lines) --")
                for ln in lines[:10]:
                    print("   ", ln)
                count += 1
    except Exception as e:
        pass
print(f"\nTotal files that look like YOLO label files (numeric lines found): {count}")

print("\n==== 2) Check for data.yaml files and print their contents (if present) ====")
data_yaml_candidates = list(pd(root).rglob('data*.yaml')) + list(pd(root).rglob('data*.yml'))
data_yaml_candidates = sorted(set([str(p) for p in data_yaml_candidates]))
if not data_yaml_candidates:
    print("No data.yaml / data.yml found under", root)
else:
    for yfile in data_yaml_candidates:
        print("\n---", yfile, "---")
        try:
            with open(yfile) as fh:
                txt = fh.read()
                print(txt[:4000])  # print up to 4000 chars
        except Exception as e:
            print("Could not read:", e)

print("\n==== 3) Inspect checkpoint (.pt) top-level keys and any embedded names/nc ====")
ckpt_path = '/content/yolov5s.pt'   # change if different
if not os.path.exists(ckpt_path):
    print("Checkpoint not found at", ckpt_path)
else:
    try:
        ck = torch.load(ckpt_path, map_location='cpu')
        if isinstance(ck, dict):
            print("Top-level keys in checkpoint:", list(ck.keys()))
            if 'names' in ck:
                print("Checkpoint 'names':", ck['names'])
            if 'nc' in ck:
                print("Checkpoint 'nc':", ck['nc'])
            if 'model' in ck:
                print("'model' type:", type(ck['model']))
                try:
                    # sometimes ck['model'] is an object; try to get attributes
                    m = ck['model']
                    if isinstance(m, dict):
                        print("model keys:", list(m.keys())[:40])
                        if 'names' in m:
                            print("model['names']:", m['names'])
                        if 'nc' in m:
                            print("model['nc']:", m['nc'])
                    else:
                        # print summary
                        print("model is not dict-like; repr:", repr(m)[:400])
                except Exception as e:
                    print("Could not inspect ck['model']:", e)
        else:
            print("Checkpoint loaded but top-level type is", type(ck))
    except Exception as e:
        print("Error loading checkpoint:", e)

print("\n==== DONE ====")
```

```python
# analyze_preds.py
import torch, yaml, os, glob, collections
from pathlib import Path
from tqdm import tqdm

# load model
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/yolov5s.pt', source='github')
print("Loaded model, len(model.names) =", len(model.names))

# load data.yaml to find val images folder
with open('/content/Rakshanetra/yolov5/data/data.yaml') as f:   # adjust if your data.yaml path is different
    data = yaml.safe_load(f)
val_images_dir = data.get('val')
if not val_images_dir:
    raise SystemExit("Could not read 'val' from data.yaml. Edit the path to your val images folder.")

# collect image file paths
img_exts = ('*.jpg','*.jpeg','*.png','*.bmp','*.webp')
img_files = []
for ext in img_exts:
    img_files += glob.glob(os.path.join(val_images_dir, ext))
img_files = sorted(img_files)
print(f"Found {len(img_files)} images in val folder: {val_images_dir}")

counts = collections.Counter()
examples = {}  # keep one example image path per predicted index

for img in tqdm(img_files, desc="Running inference"):
    results = model(img)
    if len(results.pred) and results.pred[0] is not None and results.pred[0].nelement():
        preds = results.pred[0]
        classes = preds[:, -1].cpu().numpy().astype(int).tolist()
        for c in classes:
            counts[c] += 1
            examples.setdefault(c, img)

# show summary
print("Predicted class index counts (index: count) — most common first:")
for idx, cnt in counts.most_common():
    print(f"  {idx}: {cnt}    example image: {examples.get(idx)}")

if not counts:
    print("No detections found across validation images. Try running on single images to debug.")
```

```python
# infer_and_filter.py
import torch
import os
from pathlib import Path

model = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/yolov5s.pt', source='github')

# mapping: predicted_index -> target_class_name
# update this mapping after running the analysis script if the model outputs other indices
mapping = {
    20: 'Elephant',   # COCO 20 is elephant (if your model uses 20 -> Elephant)
    21: 'Wildboar',
    15: 'Leopard',# we observed 21 -> Wildboar earlier
    # add mapping for Leopard if you observe its index (e.g., 14: 'Leopard')
}

# Build a safe names list for display (length = max_index+1)
max_idx = max(mapping.keys())
names = [model.names[i] if i < len(model.names) else f'class_{i}' for i in range(max_idx+1)]
for idx, label in mapping.items():
    names[idx] = label
model.names = names

# Function to run inference and filter results to only mapped classes
def predict_and_keep_only_mapped(img_path, conf_thresh=0.25):
    results = model(img_path)
    # results.pred[0] is tensor Nx6 [x1,y1,x2,y2,conf,class]
    if not (len(results.pred) and results.pred[0] is not None and results.pred[0].nelement()):
        return []  # no detections

    pred = results.pred[0].cpu()
    keep_rows = []
    for row in pred:
        conf = float(row[4])
        cls_idx = int(row[5].item())
        if conf < conf_thresh:
            continue
        if cls_idx not in mapping:
            continue  # discard predictions not mapped to your 3 classes
        # create a readable dict
        keep_rows.append({
            'bbox': [float(row[0]), float(row[1]), float(row[2]), float(row[3])],
            'conf': conf,
            'class_idx': cls_idx,
            'label': mapping[cls_idx]
        })
    return keep_rows

# Example usage
img = '/content/WhatsApp Image 2025-09-04 at 22.32.09_1611599c.jpg'
kept = predict_and_keep_only_mapped(img, conf_thresh=0.25)
print("Kept detections:", kept)

# If you want to save annotated image with only kept labels, render on top of the image:
from PIL import Image, ImageDraw, ImageFont
def save_annotated(img_path, detections, out_path='annotated.png'):
    im = Image.open(img_path).convert("RGB")
    draw = ImageDraw.Draw(im)
    for d in detections:
        x1, y1, x2, y2 = d['bbox']
        draw.rectangle([x1,y1,x2,y2], width=2)
        draw.text((x1, y1-12), f"{d['label']} {d['conf']:.2f}")
    im.save(out_path)
    print("Saved", out_path)

if kept:
    save_annotated(img, kept, out_path='runs/detect/filtered_result.png')
else:
    print("No mapped detections to save.")
```

```python
# infer_and_filter_updated.py
import torch
import os
from pathlib import Path
from PIL import Image, ImageDraw

# load model
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/yolov5s.pt', source='github')

# mapping: predicted_index -> target_class_name
mapping = {
    20: 'Elephant',   # COCO 20 -> Elephant
    21: 'Wildboar',   # COCO 21 -> Wildboar (your observed mapping)
    23: 'Leopard',    # observed for your leopard image (results.pred showed 23)
    # add any others as you discover them
}

# Build a safe names list for display (length = max_index+1)
max_idx = max(mapping.keys())
names = [model.names[i] if i < len(model.names) else f'class_{i}' for i in range(max_idx+1)]
for idx, label in mapping.items():
    names[idx] = label
model.names = names

# Function to run inference, keep only mapped classes, and log unknown indices
def predict_and_keep_only_mapped(img_path, conf_thresh=0.25, verbose=True):
    results = model(img_path)
    # results.pred[0] is tensor Nx6 [x1,y1,x2,y2,conf,class]
    if not (len(results.pred) and results.pred[0] is not None and results.pred[0].nelement()):
        if verbose: print("No detections.")
        return []

    pred = results.pred[0].cpu()
    keep_rows = []
    unknown_indices = set()
    for row in pred:
        conf = float(row[4])
        cls_idx = int(row[5].item())
        if conf < conf_thresh:
            continue
        if cls_idx not in mapping:
            unknown_indices.add(cls_idx)
            continue  # discard predictions not mapped to your 3 classes
        # create a readable dict
        keep_rows.append({
            'bbox': [float(row[0]), float(row[1]), float(row[2]), float(row[3])],
            'conf': conf,
            'class_idx': cls_idx,
            'label': mapping[cls_idx]
        })

    if verbose:
        if unknown_indices:
            print("Found unknown predicted indices (not in mapping):", sorted(list(unknown_indices)))
            print("Add them to `mapping` to keep/rename those predictions.")
        print("Kept detections:", keep_rows)
    return keep_rows

# Helper to save annotated image with only kept labels
def save_annotated(img_path, detections, out_path='runs/detect/filtered_result.png'):
    os.makedirs(os.path.dirname(out_path), exist_ok=True)
    im = Image.open(img_path).convert("RGB")
    draw = ImageDraw.Draw(im)
    for d in detections:
        x1, y1, x2, y2 = d['bbox']
        draw.rectangle([x1,y1,x2,y2], width=2)
        draw.text((x1, max(0, y1-14)), f"{d['label']} {d['conf']:.2f}")
    im.save(out_path)
    print("Saved", out_path)

# === Example usage ===
img = '/content/WhatsApp Image 2025-09-04 at 22.32.09_1611599c.jpg'   # change to your leopard image path
kept = predict_and_keep_only_mapped(img, conf_thresh=0.25)
if kept:
    save_annotated(img, kept, out_path='runs/detect/filtered_leopard.png')
else:
    print("No mapped detections to save.")
```
